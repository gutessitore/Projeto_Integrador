```python
##Para validar se uma série é estacionária, fizemos 2 análises: 
#Uma visual através do gráfico e a Segunda através do teste Augumented Dickey_Fuller
```

```python
#estacionária
df_vazao['val_vaz_natr'].plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
x = df_vazao.val_vaz_natr.values
x
```

```python

test_adfuller = adfuller(x)
test_adfuller
```

```python
print('ADF Statistic: %f' % test_adfuller[0]) #resultado do teste
print('p-value: %f' % test_adfuller[1]) #p-valor 
print('Critical Values:') #números críticos se a hipóste é valida ou não.
for key, value in test_adfuller[4].items():
	print('\t%s: %.3f' % (key, value))
```
